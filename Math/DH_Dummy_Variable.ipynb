1.from_formula 메서드를 사용하여 보스턴 집값 예측 문제를 다시 풀어라. C()를 사용하여 카테고리값임을 명시적으로 지정하는 것을 잊으면 안된다.

2.모형 문자열에서 상수항을 없애지 말고 다시 푼다. 이때 CHAS 변수의 가중치는 무엇을 뜻하는가.

```python
from sklearn.datasets import load_boston

boston = load_boston()
dfX0_boston = pd.DataFrame(boston.data, columns=boston.feature_names)
dfX_boston = sm.add_constant(dfX0_boston)
dfy_boston = pd.DataFrame(boston.target, columns=["MEDV"])
df_boston = pd.concat([dfX_boston, dfy_boston], axis=1)
```

```python
df_boston.tail()
```

```python
model = sm.OLS.from_formula("MEDV ~ CRIM + ZN +INDUS+CHAS+NOX+RM+AGE+DIS+RAD+TAX+PTRATIO+B+LSTAT+ C(CHAS) + 0", data=df_boston)
result = model.fit()
print(result.summary())
```

```python
model = sm.OLS.from_formula("MEDV ~ CRIM + ZN +INDUS+CHAS+NOX+RM+AGE+DIS+RAD+TAX+PTRATIO+B+LSTAT+C(CHAS)", data=df_boston)
result = model.fit()
print(result.summary())
```
