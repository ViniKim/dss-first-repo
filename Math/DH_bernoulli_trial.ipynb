```python
rv = sp.stats.norm()
type(rv)
```

```python
rv = sp.stats.norm(loc=1, scale=0.2)
```

```python
xx = np.linspace(-2, 2, 100)
pdf = rv.pdf(xx)
plt.plot(xx, pdf)
plt.grid(True)
plt.show()
# scale(표준편차=폭)이 좁아졌으니까 y값이 1이 넘어가는 거(=확률밀도가 1보다 커진 거) + scale을 0.2로 했으니까 범위를 (-8,8)이 아니라
# (-2,2) 정도로 해야 함.
```

```python
# 기댓값은 1 정도. 폭은 뿌리 정도로 내려오면 표준편차의 2배나 3배 정도가 됨.
```

```python
xx = np.linspace(-8, 8, 100)
cdf = rv.cdf(xx)
plt.plot(xx, cdf)
plt.grid(True)
plt.show()
```

베르누이 확률 분포의 모수가 다음과 같을 경우에 각각 샘플을 생성한 후 기댓값과 분산을 구하고 앞의 예제와 같이 확률 밀도 함수와 비교한 카운트 플롯을 그린다. 샘플의 갯수가 10개인 경우와 1000개인 경우에 대해 각각 위의 계산을 한다.

θ=0.5 
θ=0.9

```python
theta1 = 0.5
rv1 = sp.stats.bernoulli(theta1)
```

```python
x1 = rv1.rvs(10, random_state=0)
x1
```

```python
np.mean(x1), np.var(x1, ddof=1)
```

```python
sns.countplot(x1)
plt.title("베르누이 확률변수의 시뮬레이션 결과")
plt.xlabel("표본값")
plt.grid(True)
plt.show()
```

```python
x2 = rv1.rvs(100, random_state=0)
x2
```

```python
np.mean(x2), np.var(x2, ddof=1)
```

```python
sns.countplot(x2)
plt.title("베르누이 확률변수의 시뮬레이션 결과")
plt.xlabel("표본값")
plt.grid(True)
plt.show()
```
