```python
db.getCollection('info').find({})


update

db.collection_name.update( query, update, {upsert:<bool>, multi:<bool> } )
query: mysql - where clause
update: mysql - set
upsert: insert + update (if O, update, X, insert)
multi: (default: false) 
    

db.info.update(
    { subject:"css" },
    { subject:"sass", level: 2, comments: [{name: "alice", msg:"hello"}] }
)


db.info.update(
    { subject:"python" },
    { subject:"java", level: 2, comments: [{name: "alice", msg:"hello"}] }
)
    
db.info.update(
    { subject:"less" },
    { subject:"less", level: 2, comments: [{name: "alice", msg:"hello"}] },
    { upsert:true }
)
    
db.info.update(
    { subject:"less" },
    { subject:"less", level: 3, comments: [{name: "alice", msg:"hello"}] },
    { upsert:true }
)
    
$set : amending certain field
$unset : delete
    
db.info.update(
    { subject: "java"},
    { $set: { level:4 } },
    { multi: true}
)
    
db.info.update(
    { subject: "html"},
    { $unset: { level:1 } },
    { multi: true}
)
    
// level <= 3 --> level 1
    
db.info.update(
    { level: {$lte: 3} } ,
    { $set : {level:1} } ,
    { multi: true}
)
    
// add level 3

db.info.update(
    { level: {$exists: false} } ,
    { $set : {level:3} } ,
    { multi: true}
)    
    
$push : 
$each :
    
// subject = html -> add documnet comment
    
db.info.update(
    {subject: "html"},
    {$push: {comments: {name:"peter", msg:"html is good!"} } }
)
    
db.info.update(
    {subject: "html"},
    {$push: {comments: {$each: [ {name:"jin", msg:"html is not bad!"}, { name:"alice", msg:"htiml is good"} ] } } }
)    
    
$pull:
$in:
    
db.info.update(
    {subject: "html"},
    {$pull: {comments: {name:"peter", msg:"html is good!"} } }
)
    
db.info.update(
    {subject: "html"},
    {$pull: {comments: {$in: [ {name:"jin", msg:"html is not bad!"}, { name:"alice", msg:"htiml is good"} ] } } }
)
    
    
```
