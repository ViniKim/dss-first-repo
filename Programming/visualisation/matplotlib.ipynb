##### basic line plot

```python
y = np.random.randint(1, 10, size=10)
print(y)
plt.plot(y)
plt.show()
```

```python
# x 데이터 추가
x = range(0, 100, 10)
print(list(x), y)
plt.plot(x,y)
plt.show()
```

```python
# 다중데이터 표현

y1 = np.random.randint(1, 10, size=5)
y2 = np.random.randint(1, 10, size=5)
y3 = np.random.randint(1, 10, size=5)
x = range(5)
print(list(x), y1, y2, y3)
plt.plot(x, y1, 'ro-', x, y2, 'g*-', x, y3, 'b^:')
plt.show()
```

```python
plt.plot(x, y1)
plt.plot(x, y2)
plt.plot(x, y3)
plt.show()
```

##### 범례-legend
- label, place
- loc = <- 여기 숫자에 따라서 범례의 위치가 달라짐

```python
y1 = np.random.randint(1, 10 ,size=5)
y2 = np.random.randint(1, 10, size=5)
plt.plot(y1, label="y1")
plt.plot(y2, label="y2")
plt.legend(loc=0)
plt.show()
```

```python
y1 = np.random.randint(1, 10, size=5)
y2 = np.random.randint(1, 10, size=5)
plt.plot(y1, label="y1")
plt.plot(y2, label="y2")
plt.legend(loc=1)
plt.show()
```

```python
y1 = np.random.randint(1, 10, size=5)
y2 = np.random.randint(1, 10, size=5)
plt.plot(y1, label="y1")
plt.plot(y2, label="y2")
plt.legend(loc=10)
plt.show()
```

##### title 추가
